```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

loja.head()
```

```python
# Informações básicas sobre cada loja

print("Loja 1")
loja.info()
print("\nLoja 2")
loja2.info()
print("\nLoja 3")
loja3.info()
print("\nLoja 4")
loja4.info()
```

```python
# Verificando valores nulos em cada loja

print("Valores nulos - Loja 1")
print(loja.isnull().sum())

print("\nValores nulos - Loja 2")
print(loja2.isnull().sum())

print("\nValores nulos - Loja 3")
print(loja3.isnull().sum())

print("\nValores nulos - Loja 4")
print(loja4.isnull().sum())
```

```python
print(loja.columns)
```

```python
import matplotlib.pyplot as plt

# Lista com os DataFrames das lojas
lojas = [loja, loja2, loja3, loja4]

# Nomes das lojas
nomes_lojas = ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4']

# Lista para armazenar o faturamento de cada loja
faturamentos = []

# Calculando o faturamento de cada loja
for df in lojas:
    faturamentos.append(df['Preço'].sum())

# Exibindo os faturamentos
for nome, valor in zip(nomes_lojas, faturamentos):
    print(f'{nome}: R$ {valor:,.2f}')

# Criando o gráfico de barras
plt.figure(figsize=(8,5))
plt.bar(nomes_lojas, faturamentos, color='skyblue')
plt.title('Faturamento Total por Loja')
plt.ylabel('Faturamento (R$)')
plt.xlabel('Lojas')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
# Lista com os DataFrames das lojas

lojas = [loja, loja2, loja3, loja4]
nomes_lojas = ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4']

print("Categorias mais populares por loja:\n")
for nome, df in zip(nomes_lojas, lojas):
    categoria_mais_popular = df['Categoria do Produto'].value_counts().idxmax()
    quantidade = df['Categoria do Produto'].value_counts().max()
    print(f'{nome}: {categoria_mais_popular} ({quantidade} vendas)')
```

```python
import matplotlib.pyplot as plt

# Lista com os DataFrames das lojas
lojas = [loja, loja2, loja3, loja4]
nomes_lojas = ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4']

# Criando um dicionário para armazenar as contagens das categorias de cada loja
dados_categorias = {}

# Loop para extrair as categorias de cada loja
for nome, df in zip(nomes_lojas, lojas):
    categorias = df['Categoria do Produto'].value_counts()
    dados_categorias[nome] = categorias

# Criando um DataFrame com as categorias de todas as lojas
df_categorias = pd.DataFrame(dados_categorias).fillna(0)

# Plotando
df_categorias.plot(kind='barh', figsize=(10, 7), colormap='tab10')
plt.title('Comparação das Categorias Mais Vendidas')
plt.xlabel('Quantidade de Vendas')
plt.ylabel('Categoria')
plt.gca().invert_yaxis()  # Coloca a categoria mais vendida no topo
plt.grid(axis='x', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

```python
# Lista com os DataFrames das lojas

lojas = [loja, loja2, loja3, loja4]
nomes_lojas = ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4']

print("Avaliação média dos clientes por loja:\n")
for nome, df in zip(nomes_lojas, lojas):
    media_avaliacao = df['Avaliação da compra'].mean()
    print(f'{nome}: {media_avaliacao:.2f}')
```

```python
import matplotlib.pyplot as plt

# Nomes das lojas e respectivas médias de avaliação
nomes_lojas = ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4']
avaliacoes_medias = [loja['Avaliação da compra'].mean(),
                     loja2['Avaliação da compra'].mean(),
                     loja3['Avaliação da compra'].mean(),
                     loja4['Avaliação da compra'].mean()]

# Gráfico de pontos
plt.figure(figsize=(8, 4))
plt.plot(nomes_lojas, avaliacoes_medias, 'o-', color='seagreen', markersize=8)
plt.title('Média de Avaliação por Loja')
plt.ylabel('Avaliação Média dos Clientes')
plt.ylim(3.5, 5)
plt.grid(axis='y', linestyle='--', alpha=0.5)

# Adicionando os valores nos pontos
for x, y in zip(nomes_lojas, avaliacoes_medias):
    plt.text(x, y + 0.02, f'{y:.2f}', ha='center', va='bottom', fontsize=10)

plt.tight_layout()
plt.show()


```
